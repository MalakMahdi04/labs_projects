##### Q5.C
Write a function that accepts one parameter, a list
The function should return the sum total of all numbers in the list, using a loop.
\
*Hint*: Do we want a for loop or a while loop for this kind of question?

```python
# your code here
def my_list(numbers):
    total = 0
    for number in numbers:
        total = total + number
    return total
    
```

```python
my_list_values=[3,7,1,9]

print(my_list(my_list_values))
```

#### Q6. Problem-Solving Functions

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here
def longest_string_finder(s):
    longest =""
    for word in s:
        if len(word) > len(longest):
            longest = word
    return longest
```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]

longest_string_finder(s)
```

```python
# B. Write a function to determine whether a given list contains a given value

# your code here
def search_list(ls,value):
    for number in ls:
        if number == value:
            return True
        else: return False
```

```python
ls=[34,90,120,432,647,875]

search_list(ls,900)
```

```python
ls=[34,90,120,432,647,875]

search_list(ls,122)
```

```python
# D. Write a function to count how many times a given string contains a given letter

# your code here
def char_count(s2, character):
    count = 0
    for char in s2:
        if char == character:
            count += 1
    return count
```

```python
s2="This is a test about data science short course with excellent students"

char_count(s2,"i")
```
